```python
# Extract operator and kpi columns
df_melt["operator"] = df_melt["operator_kpi"].str.split("_").str[0]
df_melt["kpi"] = df_melt["operator_kpi"].str.split("_").str[1]
```

```python
# Pivot Dataframe to get KPI's as separate column
df_pivot = df_melt.pivot(index = ['Barangay PSGC','Barangay Name','Town PSGC','Town Name','Province','operator'], columns="kpi", values="value").reset_index()


df_pivot.head(5)
```
